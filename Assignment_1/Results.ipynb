## UnPickling the model


unpickling the models

```python
mlp_model_lr0_1 = pickle.load(open(mlp_filepath + "_lr0.1", 'rb'))
mlp_model_aug_lr0_1 = pickle.load(open(mlp_aug_filepath + "_lr0.1", 'rb'))

err_lr0_1 = pickle.load(open('TrainedModels/etc/error_vs_epoch_lr0.1', 'rb'))
err_aug_lr0_1 = pickle.load(open('TrainedModels/etc/error_vs_epoch_aug_lr0.1', 'rb'))

mlp_model_lr0_01 = pickle.load(open(mlp_filepath + "_lr0.01", 'rb'))
mlp_model_aug_lr0_01 = pickle.load(open(mlp_aug_filepath + "_lr0.01", 'rb'))

err_lr0_01 = pickle.load(open('TrainedModels/etc/error_vs_epoch_lr0.01', 'rb'))
err_aug_lr0_01 = pickle.load(open('TrainedModels/etc/error_vs_epoch_aug_lr0.01', 'rb'))

mlp_model_lr0_001 = pickle.load(open(mlp_filepath + "_lr0.001", 'rb'))
mlp_model_aug_lr0_001 = pickle.load(open(mlp_aug_filepath + "_lr0.001", 'rb'))

err_lr0_001 = pickle.load(open('TrainedModels/etc/error_vs_epoch_lr0.001', 'rb'))
err_aug_lr0_001 = pickle.load(open('TrainedModels/etc/error_vs_epoch_aug_lr0.001', 'rb'))

svm_model = pickle.load(open(svm_filepath, 'rb'))
svm_model_aug = pickle.load(open(svm_aug_filepath, 'rb'))

knn_model = pickle.load(open(knn_filepath, 'rb'))
knn_model_aug = pickle.load(open(knn_aug_filepath, 'rb'))

logistic_model = pickle.load(open(logistic_filepath, 'rb'))
logistic_model_aug = pickle.load(open(logistic_aug_filepath, 'rb'))

dtc_model = pickle.load(open(dtc_filepath, 'rb'))
dtc_model_aug = pickle.load(open(dtc_aug_filepath, 'rb'))
```

error vs epoch graph

```python
# plotting the error vs epochs graph
plt.plot(err_lr0_001)
plt.plot(err_lr0_01)
plt.plot(err_lr0_1)
plt.legend(['lr = 0.001', 'lr = 0.01', 'lr = 0.1'])
plt.xlabel('epochs')
plt.ylabel('error')
plt.ylim(0, 0.005)
plt.title('error vs epochs for original data')
plt.show()

# plotting the error vs epochs graph
plt.plot(err_aug_lr0_001)
plt.plot(err_aug_lr0_01)
plt.plot(err_aug_lr0_1)
plt.xlabel('epochs')
plt.legend(['lr = 0.001', 'lr = 0.01', 'lr = 0.1'])
plt.ylabel('error')
plt.title('error vs epochs for augmented data')
plt.show()
```

# Training The Machine Learning models Using the scikit learn package

1. SVM

```python
from sklearn import metrics

# predicting
svm_predict = svm_model.predict(data.x_test)
svm_predict_aug = svm_model_aug.predict(data.x_test_aug)

acc_svm_train = metrics.accuracy_score(data.y_train_raw, svm_model.predict(data.x_train))
acc_svm_test = metrics.accuracy_score(data.y_test_raw, svm_predict)

print("Train set Accuracy: ", acc_svm_train)
print("Test set Accuracy: ", acc_svm_test)

acc_svm_train_aug = metrics.accuracy_score(data.y_train_raw_aug, svm_model_aug.predict(data.x_train_aug))
acc_svm_test_aug = metrics.accuracy_score(data.y_test_raw, svm_predict_aug)

print("Augmented Train set Accuracy: ", acc_svm_train_aug)
print("Augmented Test set Accuracy: ", acc_svm_test_aug)
```

2. KNN Classifier

```python
from sklearn import metrics

# prediction
knn_predict = knn_model.predict(data.x_test)
knn_predict_aug = knn_model.predict(data.x_test_aug)

# calculating the accuracy
acc_knn_train = metrics.accuracy_score(data.y_train, knn_model.predict(data.x_train))
acc_knn_test = metrics.accuracy_score(data.y_test, knn_predict)

print("Train set Accuracy: ", acc_knn_train)
print("Test set Accuracy: ", acc_knn_test)

acc_knn_train_aug = metrics.accuracy_score(data.y_train_aug, knn_model_aug.predict(data.x_train_aug))
acc_knn_test_aug = metrics.accuracy_score(data.y_test, knn_predict_aug)

print("Augmented Train set Accuracy: ", acc_knn_train_aug)
print("Augmented Test set Accuracy: ", acc_knn_test_aug)
```

3. Logistic Regression Classifier

```python
from sklearn import metrics

# prediction
logistic_predict = logistic_model.predict(data.x_test)
logistic_predict_aug = logistic_model_aug.predict(data.x_test_aug)

# calculating the accuracy
acc_logistic_train = metrics.accuracy_score(data.y_train_raw, logistic_model.predict(data.x_train))
acc_logistic_test = metrics.accuracy_score(data.y_test_raw, logistic_predict)

print("Train set Accuracy: ", acc_logistic_train)
print("Test set Accuracy: ", acc_logistic_test)

acc_logistic_train_aug = metrics.accuracy_score(data.y_train_raw_aug, logistic_model_aug.predict(data.x_train_aug))
acc_logistic_test_aug = metrics.accuracy_score(data.y_test_raw, logistic_predict_aug)

print("Augmented Train set Accuracy: ", acc_logistic_train_aug)
print("Augmented Test set Accuracy: ", acc_logistic_test_aug)
```

4. Descision Tree Classifier

```python
from sklearn import metrics

# predicting the model
dtc_predict = dtc_model.predict(data.x_test)
dtc_predict_aug = dtc_model_aug.predict(data.x_test_aug)

# calculating the accuracy
acc_dtc_train = metrics.accuracy_score(data.y_train, dtc_model.predict(data.x_train))
acc_dtc_test = metrics.accuracy_score(data.y_test, dtc_predict)

print("Train set Accuracy: ", acc_dtc_train)
print("Test set Accuracy: ", acc_dtc_test)

acc_dtc_train_aug = metrics.accuracy_score(data.y_train_aug, dtc_model_aug.predict(data.x_train_aug))
acc_dtc_test_aug = metrics.accuracy_score(data.y_test, dtc_predict_aug)

print("Augmented Train set Accuracy: ", acc_dtc_train_aug)
print("Augmented Test set Accuracy: ", acc_dtc_test_aug)
```
